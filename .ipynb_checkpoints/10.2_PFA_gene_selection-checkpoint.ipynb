```python
from principal_feature_analysis import pfa 
import pandas as pd
```

```python
res,mutual_information=pfa(path="/home2/lumpi/Projects/single-cell_samantha/data/PFA_analysis_data.csv", calculate_mutual_information=True,
                           cluster_size=200, min_n_datapoints_a_bin=50)
mutual_information=mutual_information[0]

gene_names=pd.read_csv('/home2/lumpi/Projects/single-cell_samantha/data/gene_names.csv')
gene_names=gene_names.iloc[res,:].reset_index(drop=True)


mutual_information=mutual_information.join(gene_names)

mutual_information_sorted=mutual_information.sort_values(by=["mutual information"], ascending=False)
mutual_information_sorted.to_csv('/home2/lumpi/Projects/single-cell_samantha/data/gene_mutual_information.csv', index=False)

f = open("principal_features_for_model.txt", "w")
for i in res:
    f.write(str(i)+str(","))
f.close()
```
